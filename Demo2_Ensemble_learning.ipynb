## Load iris dataset from sklean in built data

```python
from sklearn.datasets import load_iris
```

## Load iris data from sklearn in built datasets

```python
iris_data = load_iris()
print(iris_data)
```

## Extract both data values and target values

```python
data_input = iris_data.data
data_output = iris_data.target
```

```python
print(data_input)
```

```python
print(data_output)
```

## Load KFold package and define the parameters

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True)
```

## Iterate through data input and output train and test sets

```python
print("Train Set     Test Set      ")
for train_set,test_set in kf.split(data_input):
    print(train_set,test_set)
```

## Load and initialize Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier(n_estimators=10)
```

## Import cross validation score from sklearn to evaluate model performance 

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(rf_class,data_input,data_output,scoring="accuracy",cv=10))
```

## Check model accuracy

```python
acc = cross_val_score(rf_class,data_input,data_output,scoring="accuracy",cv=10).mean()*100
print("Accuracy of Random Forest is: ",acc)
```
